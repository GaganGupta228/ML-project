```python
import  pickle
import numpy as np

with open ('/home/gagan/Downloads/train_image.pkl','rb') as f:
    X=pickle.load(f)
X=np.array(X)

with open('/home/gagan/Downloads/train_label.pkl','rb') as f:
    Y=pickle.load(f)
Y = np.array(Y)
```

```python
l=[]      # Im doing this because the classes are given as 0,2,3,6 which might cause problem in solving the problem
          #I m marking class with sequenced number i.e 0,1,2,3   0-0 ,2-1,3-2,6-4
```

NAIVE BAYES CLASSIFIER
- Multinomial
- Gaussian

- We will how bayes classifier works on our given data set

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,MultinomialNB

mnb=MultinomialNB()
gnb=GaussianNB()

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)

#Train

mnb.fit(x_train,y_train)
gnb.fit(x_train,y_train)

#Score

print(mnb.score(x_test,y_test))
print(gnb.score(x_test,y_test)) 
```

```python
print(y_train.shape) 
print(Y)## we can see that data is very well shuffled by test train split
```

So the performance of Multinomial is quite better than Gaussian because
- We can treat the pixel of image as discrete feature or value for the Image
- While gaussian take continous feature which is not that much true in case of pixels

SVM Classifier
- SVM classifier can also be used in Image classification
- we will be using inbuilt svm model which are by Default REST Vs ONE
- We will be checking accuracy for different kernels

```python
from sklearn import svm

#Train

svr=svm.SVC()   #RBF by Default

svr.fit(x_train,y_train)
print(svr.score(x_test,y_test))

svp=svm.SVC(kernel='poly')      #kernel is using polynomial Transformation
svp.fit(x_train,y_train)
print(svp.score(x_test,y_test))


svs=svm.SVC(kernel='sigmoid')
svs.fit(x_train,y_train)
print(svs.score(x_test,y_test))
```

It is clearly seen that SVM with polynomial Kernel has very good accuracy on Data set while that of other is very poor

CNN Network

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Input,Convolution2D,MaxPooling2D,Flatten,Dense,Dropout
from keras.utils import np_utils
import tensorflow
```

Data Preperation

```python
x_train=x_train.reshape((-1,28,28,1))
```

```python
print(np.unique(y_train,return_counts=True))
y_train= np_utils.to_categorical(y_train)
print(y_train.shape)
print(x_train.shape)
```

Model Architecture

```python
model=Sequential()
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(Convolution2D(64,(3,3),activation='relu'))
model.add(Dropout(0.25))
model.add(MaxPooling2D(2,2))


model.add(Convolution2D(8,(3,3),activation='relu'))
model.add(Convolution2D(8,(3,3),activation='relu'))

model.add(Flatten())
model.add(Dense(4,activation='softmax'))
```

```python
model.summary()
```

Complile The Neural Network

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

Training 

```python
hist=model.fit(x_train,y_train,epochs=20,shuffle=False,batch_size=32,validation_split=0.20)
```

```python
print(max(hist.history['acc']))
```

So CNN Network Has produced Highest accuracy 0f 91 which can be furture increased with no of iteration

So List of accuracy Goes like this:
    - CNN +90
    - SVM with poly kernel 82%
    -Multinomial Naive bayes 72%
